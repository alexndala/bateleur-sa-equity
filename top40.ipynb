import libraries


```python
import numpy as np 
import pandas as pd 
import yfinance as yf
import seaborn as sns
import matplotlib as mpl 
import matplotlib.pyplot as plt

plt.style.use('ggplot')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# download the data 

start = '2004-11-30'
end = '2024-11-30'

df = yf.download('^J203.JO', start=start, end=end, interval='1mo')
# df = df.loc[:,('Adj Close')]

# print(df)
```

This notebook provides a comprehensive visual analysis of the FTSE/JSE Top 40 index's monthly performance.

In this notebook, we present a heatmap that illustrates the monthly changes in the index over a historical period. The heatmap allows us to quickly identify trends, patterns, and seasonal variations in the index's performance.

Aim: 

Visualize how the index has performed over different months and years.
- Identify months with significant positive or negative changes.
- Analyze the historical volatility and seasonality of the index.


```python
# Reset the index to make 'Date' a column  
df.reset_index(inplace=True)  

# Ensure the 'Date' column is in datetime format  
df['Date'] = pd.to_datetime(df['Date'])  

# Extract year and month  
df['Year'] = df['Date'].dt.year  
df['Month'] = df['Date'].dt.month  

# Calculate monthly returns 
df['Monthly_Return'] = df['Close'].pct_change()  

# Group by year and month to calculate average monthly performance  
monthly_performance = df.groupby(['Year', 'Month'])['Monthly_Return'].mean().unstack()  

# Plot the heatmap  
plt.figure(figsize=(12, 8))  
sns.heatmap(monthly_performance, cmap='coolwarm', annot=True, fmt=".2%")  
plt.title('Copyright (c) 2024, Mahlatse A. Ndala. All rights reserved.\n FTSE/JSE Top 40 Index Monthly Performance')
plt.xlabel('Month')
plt.ylabel('Year') 
plt.show()
```
